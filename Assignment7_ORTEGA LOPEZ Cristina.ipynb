```python
#preprocessing
import re
import nltk
import string
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = text.lower()
    text = re.sub(r'\S*@\S*\s?', '', text)  #removing emails
    text = re.sub(r'\s+', ' ', text)        #extra whitespace
    text = re.sub(r"[^a-zA-Z]", " ", text)  #only letters
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words and len(word) > 3]
    return " ".join(tokens)

cleaned_docs = [clean_text(doc) for doc in newsgroup_data]
```

```python
#LDA technique
count_vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
count_data = count_vectorizer.fit_transform(cleaned_docs)

#NMF technique
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')
tfidf_data = tfidf_vectorizer.fit_transform(cleaned_docs)
```

```python
from sklearn.decomposition import LatentDirichletAllocation, NMF

n_topics = 10

lda = LatentDirichletAllocation(n_components=n_topics, random_state=0)
lda_topics = lda.fit_transform(count_data)

nmf = NMF(n_components=n_topics, random_state=1)
nmf_topics = nmf.fit_transform(tfidf_data)
```

```python
#top words per topic
import matplotlib.pyplot as plt
import seaborn as sns

def plot_top_words(model, feature_names, n_top_words, title):
    fig, axes = plt.subplots(2, 5, figsize=(15, 7))
    axes = axes.flatten()
    for topic_idx, topic in enumerate(model.components_):
        top_features = [feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]
        weights = topic[topic.argsort()[:-n_top_words - 1:-1]]
        ax = axes[topic_idx]
        sns.barplot(x=weights, y=top_features, ax=ax)
        ax.set_title(f'Topic {topic_idx + 1}')
        ax.set_xlabel('Weight')
    fig.suptitle(title, fontsize=16)
    plt.tight_layout()
    plt.show()

# LDA
plot_top_words(lda, count_vectorizer.get_feature_names_out(), 10, 'LDA Topics')

# NMF
plot_top_words(nmf, tfidf_vectorizer.get_feature_names_out(), 10, 'NMF Topics')
```

```python
from wordcloud import WordCloud

def draw_wordcloud(model, feature_names, title):
    for topic_idx, topic in enumerate(model.components_):
        topic_words = {feature_names[i]: topic[i] for i in topic.argsort()[:-30:-1]}
        wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(topic_words)
        plt.figure(figsize=(8, 4))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title(f'{title} - Topic {topic_idx + 1}')
        plt.show()

# LDA
draw_wordcloud(lda, count_vectorizer.get_feature_names_out(), "LDA")

# NMF
draw_wordcloud(nmf, tfidf_vectorizer.get_feature_names_out(), "NMF")
```
